```python
def driving(car_type,mileage,miles_per_year,energy_of_country,intensity_of_country): #lbs per year
    #car type refers to electric, gas, or diesel. If electric, mileage must become range
    #mileage must be in miles per gallon
    
    
    #Hybrids are just gas cars with a higher mpg in this type of calculation
    if car_type == 'Hybrid':
        car_type == 'Gas'
    gas_constant = 19.07 #source 4 lbs/gallon
    diesel_constant = 22.44 #source 5
    electric_constant = 3.5 #reviewed 20 different electric cars and this seemed to be the average miles/kwh
    # check units
    if car_type == "Electric":
        emissions = electricity(miles_per_year/electric_constant,energy_of_country,intensity_of_country)
    elif car_type == "Gas":
        emissions = miles_per_year * gas_constant / mileage  #need to check units out
    else:
        emissions = miles_per_year * diesel_constant / mileage  #need to check units out
    return emissions
```

```python
def car_manufacturing(car_type,new): #lbs per year
    #this is really hard to calculate and should be taken with a grain of salt. There is a lot of factors
    #that go into this and adding them together is difficult. Therefore, finding trustworthy data is nearly
    #impossible. Therefore, I used source 10, which references a website that I found had been referenced a few times.
    
    #Ultimately, as this is not super reliable, I think that we can be comfortable with ball-parking this figure.
    
    #I will be assuming that the carbon costs go entirely to the first buyer. Whenever you buy used, you are reusing,
    #so I don't think you should have any of the emissions associated with production.
    if new == 0:
        return 0
    tons_to_lbs = 2000
    if car_type == "Electric":
        return 8.8 * tons_to_lbs #carbon equivalent in lbs
    if car_type == "Gas" or car_type == "Diesel":
        return 5.6 * tons_to_lbs
    if car_type == "Hybrid":
        return 7.2 * tons_to_lbs #literally just assumed that a hybrid is half way between a gas car and an electric car
```

```python
def food_emission(diet): #lbs per year
    diet_type = {
        'Vegetarian': 3400,
        'Vegan': 3000,
        'Meat-heavy':6600,
        'Average':5000,
        'No beef':3800
        }
    return diet_type[diet]
```

```python
def electricity(kwh,energy_of_country,intensity_of_country): #lbs per year
    #kwh needs to be annual
    #square meter refers to home size
    #fuel types are decimals of fuel used for energy
    #car determines if we need to account for energy to charge car
    #units are kg co2 per kwh
        
    emissions = 0
    kg_to_lbs = 2.20462
    
    if kwh == "":
        #get the average energy used per capita for the country
        emissions = intensity_of_country * energy_of_country / 1000 * kg_to_lbs
    else:
        emissions = (kwh * intensity_of_country / 1000 * kg_to_lbs)
    return emissions
```

```python
def flight_habits(flights_per_year,length,class_type,roundtrip,country,mode): #lbs per year
    #roundtrip is 2 if yes, and 1 if no
    seat = f"{length}_{class_type}"

    #kg per passenger per kilometer, CO2 equivalent with radiative force (radiative forcing is source 8)
    seat_type = {
        'ss_':0.24587,
        'short_business':0.22652,
        'short_economy':0.15102,
        'long_economy':0.14787,
        'long_premium_economy':0.23659,
        'long_business':0.42882,
        'long_first':0.59147,
    }
    
    #in kilometers
    #source 9 provides above numbers
    #international should probably be higher, but the site didn't provide it so low-balling is the best bet    
    flight_distance = {
        'ss':415,
        'short':1316,
        'long':6924,
    }

    
    kg_to_lbs = 2.20462
    if mode == 1:
        emissions = flights_per_year * flight_distance[length] * seat_type[seat] * kg_to_lbs * roundtrip
    else:
        #this file accounts for the fact that the passenger may not be a citizen from that country
        #i.e. this accurately shows the footprint of individuals who live in that country
        file_df2 = pd.read_csv('per-capita-co2-aviation-adjusted.csv')
        flight_file = file_df2.to_numpy()
        average_co2_per_capita_year = {}
        for i in range(len(energy_file)):
            average_co2_per_capita_year.update({flight_file[i][0]:flight_file[i][3]})
        if country in average_co2_per_capita_year:
            emissions = average_co2_per_capita_year[country]
        else:
            emissions = average_co2_per_capita_year['World']
    return emissions 
```

```python
def rail(miles_per_year): #lbs per year
    miles_to_km = 1.60934
    rail_constant = 0.00446 # source 9
    kg_to_lbs = 2.20462

    return miles_per_year * miles_to_km * kg_to_lbs * rail_constant
```
